### Handling null values

```python
df = pd.read_csv('orders.csv', na_values=['Not Available', 'unknown'])
df['Ship Mode'].unique()
```

### Rename column names like, lower case and replacing space with underscore(_)

```python
df.columns
```

```python
df.columns = df.columns.str.lower()
df.columns
```

```python
df.columns = df.columns.str.replace(' ', '_')
df.columns
```

```python
# To Check
df.head(5)
```

## Deriving new columns ==> discount, sale_price, profit.

#### 1. discount

```python
df['discount'] = df['list_price'] * (df['discount_percent'] * 0.01)
```

```python
df.head(5)
```

#### 2. sale_price

```python
df['sale_price'] = df['list_price'] - df['discount']
df.head(5)
```

#### 3. profit

```python
df['profit'] = df['sale_price'] - df['cost_price']
df.head(5)
```

## converting the datatype of order_date from object to date-time

```python
df['order_date'] = pd.to_datetime(df['order_date'], format = "%Y-%m-%d")
```

## Dropping columns ==> cost_price, list_price and discount_percent

```python
df.drop(columns = ['cost_price','list_price','discount_percent'], inplace = True)
```
